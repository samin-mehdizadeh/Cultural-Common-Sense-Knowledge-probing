```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sn

plt.style.use("ggplot")
```

```python
df = pd.read_csv("en_data.csv")
df
```

```python
fig = px.histogram(df, x="label", height=400, width=500, title="Labels Distribution", text_auto=True).update_xaxes(categoryorder='total ascending')
fig.show(renderer="png")
fig.write_image("labels.pdf")
```

```python
fig = px.histogram(df, y="context", height=500, title="Contexts Distribution", text_auto=True).update_yaxes(categoryorder='total ascending')
fig.show(renderer="png")
fig.write_image("contexts.pdf")
```

```python
fig = px.histogram(df, x="country", height=400, width=500, title="Countries Distribution", text_auto=True)
fig.show(renderer="png")
fig.write_image("countries.pdf")
```

```python
fig = px.histogram(df, y="GPT-3 Pred (T/F/I)", height=400, width=500, title="GPT-3 Pred (T/F/I) Distribution", text_auto=True).update_yaxes(categoryorder='total ascending')
fig.show(renderer="png")
fig.write_image("GPT3.pdf")
```
